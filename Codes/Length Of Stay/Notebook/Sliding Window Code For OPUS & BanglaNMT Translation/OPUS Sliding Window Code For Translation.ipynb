```python
#train=pd.read_csv('/kaggle/input/los-mimic/LOS/LOS_WEEKS_adm_train.csv')

#val=pd.read_csv('/kaggle/input/los-mimic/LOS/LOS_WEEKS_adm_val.csv')

test=pd.read_csv('/kaggle/input/los-of-stay/LOS/LOS_WEEKS_adm_test.csv')

#train['text']=train['text'].str.lower()

#val['text']=val['text'].str.lower()

test['text']=test['text'].str.lower()

import torch

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

print(f" Using {device}")

from transformers import pipeline



pipe = pipeline("translation", model="shhossain/opus-mt-en-to-bn",do_sample=False,use_fast=True,device='cuda')

```

```python
from tqdm import tqdm

final=[]

print('Test 4000:6000')

num=0

for text in tqdm(test['text'][4000:6000]): #text=train['text'][0:3]

    num+=1

    print(num)

    initial=[]

    start=0

    finish=300

    for i in range(9999999999):

        translation=pipe(text[start:finish],num_workers=-1)[0]['translation_text']

        if translation=='▪ (w-BE)':

            final.append(" ".join(initial))

            break

        else:

            initial.append(translation)

            start=finish

            finish=finish+300

len(final)



trans=pd.DataFrame()

trans['machine']=final

trans.to_csv('Test 4000:6000 Machine Translated  LOS')
```
