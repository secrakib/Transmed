```python
train=pd.read_csv('/kaggle/input/mydata/LOS_WEEKS_adm_train.csv')
val=pd.read_csv('/kaggle/input/mydata/LOS_WEEKS_adm_val.csv')
test=pd.read_csv('/kaggle/input/mydata/LOS_WEEKS_adm_test.csv')
train['text']=train['text'].str.lower()
val['text']=val['text'].str.lower()
test['text']=test['text'].str.lower()
```

```python
import torch
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f" Using {device}")
```

```python
from transformers import pipeline

pipe = pipeline("translation", model="csebuetnlp/banglat5_nmt_en_bn",do_sample=False,use_fast=True,device=device)
```

```python
from tqdm import tqdm
final=[]
print('range 18000:22000')             
count=0
for text in tqdm(train['text'][18000:22000]): 
    initial=[]
    count+=1
    print(count)
    start=0
    finish=800                  
    for i in range(9999999999):
        translation=pipe(text[start:finish],max_length=1000 ,num_workers=-1)[0]['translation_text']
        if translation=='কিন্তু, তা সত্ত্বেও তিনি যিহোবার প্রতি অনুগত ছিলেন।':
            final.append(" ".join(initial))
            break
        else:
            initial.append(translation)
            start=finish
            finish=finish+800    
len(final)

trans=pd.DataFrame()
trans['machine']=final
trans.to_csv('BanglaNMT Train 18000:22000 LOS')
```
