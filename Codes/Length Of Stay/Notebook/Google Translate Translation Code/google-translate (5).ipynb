```python
from tqdm import tqdm
final=[]
print("Google Train")
num=0
for text in tqdm(train['text'][0:]): #text=train['text'][0:3]
    from googletrans import Translator
    translator = Translator()
    num+=1
    print(num)
    initial=[]
    start=0
    finish=5000
    for i in range(9999999999):
        try:
            translation=translator.translate(text, src='en', dest='bn')
            final.append(translation.text)
            print('no Execpt')
            break
        except TypeError:
            print('first Execpt')
            try:
                translation=translator.translate(text[start:finish], src='en', dest='bn')
                initial.append(translation.text)
                start=finish
                finish=finish+5000
            except TypeError:
                print('2nd Execpt')
                final.append(" ".join(initial))
                break
        #translation=pipe(text[start:finish],num_workers=-1)[0]['translation_text']
        #if translation=='▪ (w-BE)':
        #else:
            #initial.append(translation)
            #start=finish
            #finish=finish+632
len(final)

trans=pd.DataFrame()
trans['machine']=final
trans.to_csv('Google Translated Train.csv',index=False)
```
