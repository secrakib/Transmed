```python
all=pd.read_csv('/kaggle/input/1000-instances/1000_instances_to_translate.csv',usecols=['text','los_label'])
```

```python
all
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(
    all, all['los_label'], test_size=0.2, random_state=0,stratify=all['los_label'])
```

```python
len(x_train) , len(x_test)
```

```python
x_train.los_label.value_counts(normalize=True)*100
```

```python
x_val, x_testi, y_val, y_testi = train_test_split(
    x_test, x_test['los_label'], test_size=0.5, random_state=0,stratify=x_test['los_label'])
```

```python
print(x_val.los_label.value_counts(normalize=True)*100,end="\n \n")
print(x_testi.los_label.value_counts(normalize=True)*100)
```
